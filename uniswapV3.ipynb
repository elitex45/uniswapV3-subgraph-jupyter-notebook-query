```python
import requests
```

```python
api_url = "https://api.studio.thegraph.com/query/51221/uniswap-v3-arbitrum/version/latest"
graphql_query = """
query MyQuery {
  positions(first: 1000, orderDirection: asc) {
    amountCollectedUSD
    amountDepositedUSD
    token0 {
      symbol
    }
    token1 {
      symbol
    }
    transaction {
      mints {
        timestamp
        tickLower
        tickUpper
      }
      burns {
        timestamp
      }
    }
    pool {
      tick
      feeTier
    }
  }
}
"""

# Define headers (optional)
headers = {
    "Content-Type": "application/json",
}

response = requests.post(api_url, json={"query": graphql_query}, headers=headers)
```

```python
def getProfitAndLoss(dataF):
    userPnL = []
    useFullData = []
    for i in range(len(dataF['data']['positions'])):
        # print(dataF)
        data = dataF
        amountCollected = float(data['data']['positions'][i]['amountCollectedUSD'])
        amountDeposited = float(data['data']['positions'][i]['amountDepositedUSD'])
        PnL = amountCollected - amountDeposited
        mintTime = data['data']['positions'][i]['transaction']['mints'][0]['timestamp']
        if data['data']['positions'][i]['transaction']['burns']:
            burnTime = data['data']['positions'][i]['transaction']['burns'][0]['timestamp']
        else:
            burnTime = 0
        currentTime = 1706539327
        if burnTime:
            age = int(burnTime)-int(mintTime)
        else:
            age = int(currentTime)-int(mintTime)
        lowerTick = data['data']['positions'][i]['transaction']['mints'][0]['tickLower']
        upperTick = data['data']['positions'][i]['transaction']['mints'][0]['tickUpper']
        currentTick = data['data']['positions'][i]['pool']['tick']

        if lowerTick < currentTick and currentTick < upperTick:
            status = "In-Range"
            var = 0
        else:
            status = "Out-Of-Range"
            var = 1

        age = age / (24 * 60 * 60)
        print("PnL:",PnL)
        print("Age:",age,"days")
        print("Lower and upper tick:",lowerTick,upperTick)
        print("Status:",status)
        print("")
        
        item = [PnL,age,status,lowerTick,upperTick]
        if var == 0 and PnL > 0:
            print(status)
            useFullData.append(item)
        userPnL.append(item)
    #print("data:",data)
    # print(userPnL)
    
    # return userPnL
    first_elements = [sublist[0] for sublist in useFullData]

    # Find the lowest and highest values
    lowest_value = min(first_elements)
    highest_value = max(first_elements)
    average_value = sum(first_elements) / len(first_elements)

    print(f"Lowest Value: {lowest_value}")
    print(f"Highest Value: {highest_value}")
    print(f"Average Value: {average_value}")

    # Extract the averages of the first and second elements
    average_first_element = sum(sublist[0] for sublist in useFullData) / len(useFullData)
    average_second_element = sum(sublist[1] for sublist in useFullData) / len(useFullData)

    # Filter arrays based on conditions
    filtered_arrays = [
        sublist for sublist in useFullData
        if sublist[0] > average_first_element and sublist[1] > average_second_element
    ]
    print("Filtered arrays:",filtered_arrays)
    print("Usefull data:",useFullData)
```

```python

if response.status_code == 200:
    data = response.json()
    # Process the data as needed
    getProfitAndLoss(data)
else:
    print(f"Error: {response.status_code}, {response.text}")
```
